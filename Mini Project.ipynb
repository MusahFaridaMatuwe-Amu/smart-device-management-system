## Mini Project: Smart Device Management System

**Scenario:**
A technology company develops software to manage smart devices used in a smart home.
Each device has common information such as its name, device ID, and power status,
but different devices have their own unique capabilities. To protect the integrity
of the system, sensitive information should be encapsulated and only modified
through appropriate methods.

**Requirements:**

1. **Parent Class: SmartDevice**
   - Private attributes: `__device_id`, `__power_status`
   - Public attribute: `name`
   - Methods: `turn_on()`, `turn_off()`, `display_info()`
   - Getter/setter methods (or `@property`) to control access to private attributes

2. **Child Classes** (all inherit from `SmartDevice` and use `super()`):
   - `SecurityCamera` — attribute: `recording_status`; methods: `start_recording()`, `stop_recording()`
   - `SmartLight` — attribute: `brightness`; methods: `increase_brightness()`, `decrease_brightness()`
   - `TemperatureSensor` — attribute: `temperature`; method: `read_temperature()`

3. **Inheritance** — each child class must inherit from `SmartDevice` and use `super()` to initialize inherited attributes.

4. **Encapsulation** — `device_id` and `power_status` must not be modified directly.
   Validation examples: brightness must be between 0 and 100; device ID cannot be empty.

5. **Object Creation** — create at least one `TemperatureSensor`, one `SmartLight`,
   and one `SecurityCamera`, and demonstrate all their methods.

6. **Good Programming Practices** — meaningful names, comments, proper indentation, modular code.

7. **Menu-Driven Interface** using loops and conditionals:
   1. Display Device Information
   2. Turn Device On
   3. Turn Device Off
   4. Read Temperature
   5. Adjust Brightness
   6. Start Recording
   7. Exit

```python



class SmartDevice:
    """Parent class representing a generic smart home device."""

    def __init__(self, name, device_id):
        self.name = name                     # Public attribute
        self.__device_id = device_id         # Private attribute
        self.__power_status = False          # Private attribute (device starts OFF)

        # Validate device ID at creation time
        if not device_id:
            raise ValueError("Device ID cannot be empty.")

    # ---------- Getter and Setter for device_id (read-only after creation) ----------
    @property
    def device_id(self):
        return self.__device_id

    # ---------- Getter and Setter for power_status ----------
    @property
    def power_status(self):
        return self.__power_status

    @power_status.setter
    def power_status(self, value):
        if not isinstance(value, bool):
            print("Error: Power status must be True or False.")
            return
        self.__power_status = value

    # ---------- Common Methods ----------
    def turn_on(self):
        self.power_status = True
        print(f"{self.name} is now ON.")

    def turn_off(self):
        self.power_status = False
        print(f"{self.name} is now OFF.")

    def display_info(self):
        status = "ON" if self.power_status else "OFF"
        print(f"Name: {self.name} | Device ID: {self.device_id} | Power Status: {status}")


class SecurityCamera(SmartDevice):
    """Child class representing a smart security camera."""

    def __init__(self, name, device_id):
        super().__init__(name, device_id)    # Initialize inherited attributes
        self.recording_status = False        # Additional attribute

    def start_recording(self):
        if not self.power_status:
            print(f"Cannot start recording: {self.name} is turned OFF.")
            return
        self.recording_status = True
        print(f"{self.name} has started recording.")

    def stop_recording(self):
        self.recording_status = False
        print(f"{self.name} has stopped recording.")

    def display_info(self):
        super().display_info()
        recording = "Recording" if self.recording_status else "Not Recording"
        print(f"Recording Status: {recording}")


class SmartLight(SmartDevice):
    """Child class representing a smart light bulb."""

    def __init__(self, name, device_id, brightness=50):
        super().__init__(name, device_id)    # Initialize inherited attributes
        self.__brightness = 0
        self.brightness = brightness         # Goes through validation via setter

    @property
    def brightness(self):
        return self.__brightness

    @brightness.setter
    def brightness(self, value):
        if not (0 <= value <= 100):
            print("Error: Brightness must be between 0 and 100.")
            return
        self.__brightness = value

    def increase_brightness(self, amount=10):
        self.brightness = min(100, self.__brightness + amount)
        print(f"{self.name} brightness increased to {self.brightness}%.")

    def decrease_brightness(self, amount=10):
        self.brightness = max(0, self.__brightness - amount)
        print(f"{self.name} brightness decreased to {self.brightness}%.")

    def display_info(self):
        super().display_info()
        print(f"Brightness: {self.brightness}%")


class TemperatureSensor(SmartDevice):
    """Child class representing a smart temperature sensor."""

    def __init__(self, name, device_id, temperature=22.0):
        super().__init__(name, device_id)    # Initialize inherited attributes
        self.temperature = temperature       # Additional attribute (in Celsius)

    def read_temperature(self):
        print(f"{self.name} current temperature reading: {self.temperature}°C")
        return self.temperature

    def display_info(self):
        super().display_info()
        print(f"Temperature: {self.temperature}°C")


# ---------------------------------------------------------------------------
# Menu-Driven Interface
# ---------------------------------------------------------------------------

def display_menu():
    """Asks a single-line question that lists all menu options compactly,
    so the whole thing fits in the input pop-up without being cut off."""
    menu_prompt = (
        "1=Show Info  2=Turn On  3=Turn Off  4=Read Temp  "
        "5=Brightness  6=Recording  7=Exit -- Enter number (1-7): "
    )
    return input(menu_prompt)


def choose_device(devices):
    """Asks a single-line question listing the devices by number,
    short enough to display fully in the input pop-up."""
    device_prompt = "  ".join(
        f"{index}={device.name}" for index, device in enumerate(devices, start=1)
    )
    try:
        choice = int(input(f"{device_prompt} -- Enter device number: "))
        if 1 <= choice <= len(devices):
            return devices[choice - 1]
        print("That number doesn't match any device. Please try again.")
    except ValueError:
        print("That's not a valid number. Please try again.")
    return None


def main():
    # Create sample devices
    living_room_light = SmartLight("Living Room Light", "LGT-001", brightness=40)
    front_door_camera = SecurityCamera("Front Door Camera", "CAM-001")
    kitchen_sensor = TemperatureSensor("Kitchen Sensor", "TMP-001", temperature=24.5)

    devices = [living_room_light, front_door_camera, kitchen_sensor]

    while True:
        choice = display_menu()

        if choice == "1":
            device = choose_device(devices)
            if device:
                print(f"\nHere is the current info for {device.name}:")
                device.display_info()

        elif choice == "2":
            device = choose_device(devices)
            if device:
                device.turn_on()

        elif choice == "3":
            device = choose_device(devices)
            if device:
                device.turn_off()

        elif choice == "4":
            if kitchen_sensor.power_status:
                kitchen_sensor.read_temperature()
            else:
                print("The Kitchen Sensor is OFF. Please turn it ON first (option 2), then try again.")

        elif choice == "5":
            if not living_room_light.power_status:
                print("The Living Room Light is OFF. Please turn it ON first (option 2), then try again.")
            else:
                action = input("Brightness: type 'up' or 'down': ").lower()
                if action == "up":
                    living_room_light.increase_brightness()
                elif action == "down":
                    living_room_light.decrease_brightness()
                else:
                    print("Didn't recognize that. Please type exactly 'up' or 'down' next time.")

        elif choice == "6":
            action = input("Recording: type 'start' or 'stop': ").lower()
            if action == "start":
                front_door_camera.start_recording()
            elif action == "stop":
                front_door_camera.stop_recording()
            else:
                print("Didn't recognize that. Please type exactly 'start' or 'stop' next time.")

        elif choice == "7":
            print("Exiting Smart Device Management System. Goodbye!")
            break

        else:
            print("That's not a valid option. Please enter a number from 1 to 7.")


if __name__ == "__main__":
    main()
```
